```python
!pip install gradio transformers googletrans==4.0.0-rc1 pillow pytesseract sentencepiece --quiet
!apt-get install -y tesseract-ocr --quiet

import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM
from googletrans import Translator
from PIL import Image
import pytesseract
import torch

# Load model
model_name = "ibm-granite/granite-3.3-2b-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")

translator = Translator()

# Chatbot
def chatbot_response(message):
    inputs = tokenizer(message, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_length=300)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Translation
def translate_text(text, target_language):
    result = translator.translate(text, dest=target_language)
    return result.text

# OCR image analysis
def analyze_image(img):
    text = pytesseract.image_to_string(img)
    return text

# Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("# üè• AI Medical Prescription Assistant")

    with gr.Tabs():
        with gr.Tab("üí¨ Chatbot"):
            msg = gr.Textbox(label="Ask anything")
            out = gr.Textbox(label="Response")
            msg.submit(chatbot_response, msg, out)

        with gr.Tab("üåê Translation"):
            inp = gr.Textbox(label="Enter text")
            lang = gr.Dropdown(["en", "hi", "te", "ta", "kn", "ml", "mr", "gu", "bn"], label="Translate to")
            out2 = gr.Textbox(label="Translated text")
            gr.Button("Translate").click(translate_text, [inp, lang], out2)

        with gr.Tab("üñº Image Analysis"):
            img = gr.Image(type="pil")
            out3 = gr.Textbox(label="Extracted Text")
            gr.Button("Extract Text").click(analyze_image, img, out3)

demo.launch()
```
